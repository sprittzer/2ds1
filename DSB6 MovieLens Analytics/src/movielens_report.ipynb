Оригинальный датасет MovieLens содержит около 10 миллиона записей. Но нам была предложена сокращённая выборка, состоящяя только из 10000 фильмов. Анализируя их, мы стремимся не просто подсчитать средние значения, а понять, как формируется эмоциональная и культурная логика аудитории.
