# Введение

## Методы машинного обучения в реальной жизни:
1) **Анализ истории просмотра/покупок/прослушивания разными сервисами для составления новой ленты видео/товаров/музыкального чарта.** Для этого используются различные алгоритмы – например, коллаборативная фильтрация (один из классических методов, основанный на поиске людей с похожими результатами), двухбашенные нейронные сети (система, состоящая из двух нейроных сетей, одна из которых обучается на данных пользователя, а другая на данных о контенте, далее система ищет ближайшие векторы контента к вектору пользователя). Такие методы позволяют реализовать быстрое нахождение пользователем релевантного контента или товара, бизнесу же это приносит рост потребления, время провождения на платформе =)
2) **Распознование изображений.** Можно вспомнить огромное количество продуктов, в которые встроен данный метод. Например, озон и другие маркетплейсы используют данный метод для реализации поиска пользователем товара, похожего на тот, что был на фотографии пользователя. Системы распознования лиц в аэропортах или камерах наблюдения. В телефонах – разблокировка по лицу. Сшиты такие системы из нейросетей (CNN, ViT, MLLM, GAN и другие). Преимущества: высокая точность, скорость, обеспечение оперативной безопасности и контроля.
3) **Голосовые помощники и машинный перевод.** Распознавание естественной речи (NLP — обработка естественного языка), понимание intent (намерения пользователя), синтез ответа и перевод между языками с учетом контекста и идиом – все это встроенно во все системы ассистенов по типу Алисы, Siri, онлайн–переводчики. Применяются архитектуры типа Whisper и нейронные сети с механизмом внимания, LLM-ки. Позволяет взаимодействовать с технологиями самым интуитивным способом — голосом, а также мгновенно получать доступ к информации на любом языке, что критически важно для глобальной коммуникации и образования!!
4) **Прогнозная аналитика и предсказание спроса супермаркетами, доствками продуктов, такси-сервисами.** Какие товары куда и в каком количестве завозить, учитывая большое нужное количество факторов (погода, праздники, население, рождаемость, курс $$$ и т.д.). Такси-сервисы (например, ЯТакси) предсказывают спрос и регулируют цены (surge pricing). Банки прогнозируют риски дефолта по кредитам. Примеров масса. Компании минимизируют потери от неликвидов или, наоборот, упущенной выгоды из-за нехватки товара. В финансах это снижение рисков и более точное ценообразование.
5) **Автономные транспортные средства.** Не везде конечно) Но такое набирает популярность в развитых странах, штатах. В данном пункте что только не используется: комплексное использование компьютерного зрения (для чтения знаков, обнаружения пешеходов), сенсорных данных (лидары, радары) и глубокого обучения для принятия решений в реальном времени — от поддержания полосы до экстренного торможения.  В перспективе это может кардинально снизить аварийность, пробки и выбросы.

## Классификация задач
### Обучение с учителем (Supervised Learning)
Задачи, где модель обучается на размеченных данных (есть входные признаки X и известные правильные ответы y).
- **Прогнозирование цены на жилье.** Целевая переменная (цена) — непрерывное число => Регрессия.
- **Прогнозирование, вернёт ли клиент кредит.** Предсказание категориального исхода с двумя классами => Бинарная классификация.
- **Прогнозирование, когда пациенту потребуется принять лекарство.** Предсказание момента времени или дозировки => Регрессия.
- **Выбор, какое из доступных лекарств должен принимать пациент.** Выбор одного оптимального варианта из множества => Многоклассовая классификация.
- **Выбор сегмента клиентов для промо-коммуникации.** Учавствует покупатель в акции – да/нет => Бинарная классификация.
- **Распознавание дефектной продукции на производственной линии (по фото).** Разделение на "дефектный" / "не дефектный" => Бинарная классификация.
- **Поиск сайтов по текстовому запросу.** Оценка релевантности сайта запросу => регрессия, либо классификация...
- **Персонализированные рекомендации.** Предсказание вероятности взаимодействия пользователя с контентом на основе его истории => Регрессия (прогноз рейтинга).
- **Поиск по фото, разблокировка по лицу.** Отнесение изображения к определённому классу (товар, лицо человека) => Многоклассовая классификация.
- **Голосовые помощники и машинный перевод.**
- **Прогнозная аналитика и предсказание спроса.** Прогнозирование временных рядов.
- **Автономные транспортные средства (компонент компьютерного зрения).** Распознавание объектов (знаков, пешеходов, машин) с камер и сенсоров → Классификация / Обнаружение объектов.

### Обучение без учителя (Unsupervised Learning)
Задачи, где модель ищет скрытые структуры или закономерности в данных без заранее известных правильных ответов (есть только X).
- **Выбор сегмента клиентов для промо-коммуникации.** Группировка клиентов с похожими характеристиками для таргетирования рекламы => Кластеризация.
- **Разделение клиентов магазина на сегменты для анализа поведения.** Поиск естественных групп среди клиентов без заранее заданных категорий => Кластеризация.
- **Обнаружение аномалий в трафике сайта.** Выявление нестандартных, подозрительных паттернов в потоке запросов => Обнаружение аномалий.
- **Автономные транспортные средства.** Анализ данных с датчиков для выявления нестандартных ситуаций или неразмеченных паттернов дорожного движения => Обнаружение аномалий / Кластеризация (может использоваться на этапе исследования данных).

## Разница между многоклассовой классификацией и многометочной классификацией
Думаб, что основное различие меж многоклассовой классификацией (multiclass) и многометочной классификацией (multilabel) заключается в принципе назначения меток объектам.

Многоклассовая: один объект = один класс (взаимоисключающие варианты).
Пример: цифра на изображении (только 0 ИЛИ 1 ИЛИ 2...)

Многометочная: один объект может соответствовать нескольким классам одновременно (не взаимоисключающие).
Пример: теги фотографии ("пляж" И "закат" И "люди").

## Можно ли считать пример с ценами на жилье классификацией или задачей регрессии?
Пример с ценами на жилье — это задача регрессии, потому что нужно предсказать конкретное число (например, 250 000$), а не выбрать категорию. Цена — непрерывная величина.

НО эту задачу можно искусственно превратить в классификацию ;) Если разбить цены на диапазоны (дешёвое, среднее, дорогое) и предсказывать уже не число, а класс. Так иногда делают, чтобы упростить выводы для бизнеса. Но проблема в том, что при таком преобразовании теряется точность — модель перестаёт видеть разницу между 110 000$ и 490 000$, если обе суммы попадают в один класс "среднее". Поэтому для точного прогноза цен стоит использовать именно регрессию.

```python
test = pd.read_json('./data/test.json')
test = test[['bathrooms', 'bedrooms', 'price']]
test.describe()
```

```python
percentile_99 = np.percentile(test['price'], 99)
percentile_1 = np.percentile(test['price'], 1)
test = test[(percentile_1 < test['price']) & (test['price'] < percentile_99) & (test['bathrooms'] <= 10)]
test.describe()
```

```python
X_train = df[['bathrooms', 'bedrooms']]
y_train = df['price']
X_test = test[['bathrooms', 'bedrooms']]
y_test = test['price']
```

```python
poly = PolynomialFeatures(10)
X_train = poly.fit_transform(X_train)
X_test = poly.transform(X_test)
```

```python
X_test
```

# Тренировка моделей

Теперь нужно обучить 3 модели: линейную регрессию, дерево решений и наивную модель

```python
result_MAE = pd.DataFrame(columns=['model', 'train', 'test'])
result_RMSE = pd.DataFrame(columns=['model', 'train', 'test'])
```

## Линейная регрессия

```python
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)
y_train_pred = lin_reg.predict(X_train)
y_test_pred = lin_reg.predict(X_test)
```

```python
mae_train = mean_absolute_error(y_train, y_train_pred)
mae_test = mean_absolute_error(y_test, y_test_pred)
rmse_train = root_mean_squared_error(y_train, y_train_pred)
rmse_test = root_mean_squared_error(y_test, y_test_pred)

result_MAE.loc[len(result_MAE)] = ['linear_regression', mae_train, mae_test]
result_RMSE.loc[len(result_RMSE)] = ['linear_regression', rmse_train, rmse_test]
```

## Дерево решений

```python
reg = DecisionTreeRegressor(random_state=21)
reg.fit(X_train, y_train)
y_train_pred_dt = reg.predict(X_train)
y_test_pred_dt = reg.predict(X_test)
```

```python
mae_train_dt = mean_absolute_error(y_train, y_train_pred_dt)
mae_test_dt = mean_absolute_error(y_test, y_test_pred_dt)

rmse_train_dt = root_mean_squared_error(y_train, y_train_pred_dt)
rmse_test_dt = root_mean_squared_error(y_test, y_test_pred_dt)

result_MAE.loc[len(result_MAE)] = ['decision_tree', mae_train_dt, mae_test_dt]
result_RMSE.loc[len(result_RMSE)] = ['decision_tree', rmse_train_dt, rmse_test_dt]
```
