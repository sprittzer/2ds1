```python
import pandas as pd
import sqlite3
```

```python
conn = sqlite3.connect('../data/checking-logs.sqlite')
```

# Exercise 00 : Line chart

```python
query = """
SELECT datetime
FROM pageviews
WHERE uid LIKE 'user_%'
"""


pageviews_df = pd.read_sql(query, conn)
conn.close()
pageviews_df

pageviews_df['date'] = pd.to_datetime(pageviews_df['datetime']).dt.date

views_per_day = pageviews_df.groupby('date').size().reset_index(name='visits')
views_per_day
```

```python
ax = views_per_day.plot(x="date", y='visits', fontsize=8, figsize=(15, 8), title='Views per day')
ax.tick_params(axis='x', rotation=90)
```
