```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
```

# Exercise 05 : Boxplot

```python
df = pd.read_csv('../data/ab-test.csv')
```

```python
test_before = df[(df['group'] == 'test') & (df['time'] == 'before')]['diff']
test_after = df[(df['group'] == 'test') & (df['time'] == 'after')]['diff']
control_before = df[(df['group'] == 'control') & (df['time'] == 'before')]['diff']
control_after = df[(df['group'] == 'control') & (df['time'] == 'after')]['diff']

display(len(test_after) + len(control_before))

fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(15, 8), sharey=True)
fig.suptitle('Lag before labs deadlines. After and before the Newsfeed', fontsize=15)

data_after = [control_after, test_after]
labels_after = ['control', 'test']
axs[0].boxplot(
    data_after, tick_labels=labels_after,
    boxprops=dict(linewidth=3, color='green'),
    medianprops=dict(linewidth=2, color='red'), 
    whiskerprops=dict(color='red'),
    capprops=dict(color='red'),
    showmeans=True
)
axs[0].set_title('after', fontsize=14)
axs[0].set_xlabel('[group]')
axs[0].grid(True, axis='y')
axs[0].grid(True, axis='x')

data_before = [control_before, test_before]
labels_before = ['control', 'test']

axs[1].boxplot(data_before, tick_labels=labels_before,
                            boxprops=dict(linewidth=3, color='green'),
                            medianprops=dict(linewidth=2, color='red'), 
                            whiskerprops=dict(color='red'),
                            capprops=dict(color='red'),
                            showmeans=True)
axs[1].set_title('before', fontsize=14)
axs[1].grid(True, axis='y')
axs[1].grid(True, axis='x')
axs[1].set_xlabel('[group]')
    
plt.tight_layout()
plt.show()
```
