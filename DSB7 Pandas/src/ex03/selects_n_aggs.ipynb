## Task 3. Make the aggregations with the make and model

```python
df.groupby('Make')['Fines'].median()
```

```python
df.groupby(['Make', 'Model'])['Fines'].median()
```

```python
df.groupby(['Make', 'Model'])['Fines'].count()
```

```python
df.groupby(['Make', 'Model'])['Fines'].agg(['min', 'max'])
```

## Task 4. Make aggregations with car numbers

The car numbers, grouped by the number of fines in descending order:

```python
fines = df.groupby('CarNumber').size().sort_values(ascending=False)
fines
```

```python
df[df.index == fines.index[0]]
```

```python
fines_sum = df.groupby('CarNumber')['Fines'].sum().sort_values(ascending=False)
fines_sum
```

```python
df[df.index == fines_sum.index[0]]
```

Are there any car numbers connected to different models?

```python
counts_diff_models = df.groupby(['CarNumber'])['Model'].nunique()
counts_diff_models[counts_diff_models > 1]
```
